### Let's learn EDA


```python
ax=sns.countplot(x="Gender",data=df)
for bar in ax.containers:
    ax.bar_label(bar)
    # ax.set_xlabel("Gender",fontsize=20)
    # ax.set_ylabel("Count",fontsize=20)
    # ax.tick_params(axis="both",labelsize=10)

plt.xlabel("Gender",size=10,color="g")
plt.ylabel("Count",size=10,color="g")
plt.title("Graphs")

plt.show()
```

```python
#Total Transactions count by Gender Wise show in Bar Chart
ax=sns.countplot(x="Gender",data=df,alpha=0.5)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Gender Wise distribution",color="k",size=15)
plt.xlabel("Gender",color="g",size=12)
plt.ylabel("count",color="g",size=12)
plt.tick_params(axis="x",size=45)
plt.tick_params(axis="y",size=45)#space

plt.show()
```

```python
#Gender wise Total Sales Amount show in bar chart
gender_wise=df.groupby("Gender",as_index=False)["Amount"].sum()
ax=sns.barplot(x="Gender",y="Amount",data=gender_wise)
for bar in ax.containers:
    ax.bar_label(bar,fmt="%.0f")
plt.show()
```

```python
#Age Group wise Transactions Count in Bar Char

ax=sns.countplot(data=df,x="Age Group")
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
#Age Group wise Total Amount
df.groupby("Age Group",as_index=False)["Amount"].sum()
```

```python
Age_wise=df.groupby("Age Group",as_index=False)["Amount"].sum()
ax=sns.barplot(data=Age_wise,x="Age Group",y="Amount")
for bar in ax.containers:
    ax.bar_label(bar,fmt="%.0f")
plt.show()
    
```

```python
# Top 5 state by Total Quantity Ordered
data_1=df.groupby("State",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head()
ax=sns.barplot(data=data_1,x="State",y="Orders")
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
data_2=df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head()
print(data_2)
```

```python
# Top 5 Sates as per Total Sales Amount
data_2=df.groupby("State",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head()
ax=sns.barplot(x="State",y="Amount",data=data_2)

for bar in ax.containers:
    ax.bar_label(bar,fmt="%.0f")
plt.show()
sns.set(rc={"figure.figsize":(30,5)})
```

```python
# Profession wise Transactions count
ax=sns.countplot(x="Profession",data=df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Profession wise Transactions count",size=25,color="g")
plt.show()
```

```python
# Amount Wise Top Professions
data1=df.groupby("Profession",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(5)
print(data1)
```

```python
data1=df.groupby("Profession",as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(5)
ax=sns.barplot(data=data1,x="Profession",y="Amount")
for bar in ax.containers:
    ax.bar_label(bar,fmt="%.0f")
plt.title("Amount Wise Top Professions")
plt.show()
```

```python
product=df.groupby("Product_Category")["Orders"].sum().sort_values(ascending=False).head(10)
print(product)
```

```python
# Top 10 Product by Total Quantity Ordered
product=df.groupby("Product_Category",as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(5)
ax=sns.barplot(data=product,x="Product_Category",y="Orders")
for bar in ax.containers:
    ax.bar_label(bar)
plt.title("Top 10 Product by Total Quantity Ordered")
plt.show()
```
